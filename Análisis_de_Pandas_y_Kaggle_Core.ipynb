```python
grouped = df.groupby('Survived')
grouped.size()
```

edad Promedio de pasajeros hombres y mujeres que sobreviveron y murieron

```python
df.groupby(['Survived', 'Sex'])['Age'].mean().round(2)
```

```python
# Sobrevivientes por sexo
sobrevivientes_por_sexo = df.groupby('Survived')['Sex'].value_counts()
sobrevivientes_por_sexo
```

```python
# Sobrevivientes por clase
conteo_sobrevivientes_por_clase = df.groupby('Survived')['Pclass'].count()
conteo_sobrevivientes_por_clase
```

```python
# Sobrevivientes mínimas y máximas por clase y sexo
sobrevivientes_por_clase = df.groupby('Survived')['Pclass'].min()
sobrevivientes_por_sexo = df.groupby('Survived')['Sex'].max()
sobrevivientes_por_clase
sobrevivientes_por_sexo
```

```python
# Agrupar por 'Survived' y 'Pclass'
grouped_multi = df.groupby(['Survived', 'Pclass'])

# Suma de Age por Surived y Pclass
suma_sobrevivientes_por_clase = grouped_multi['Age'].sum()
suma_sobrevivientes_por_clase

# Promedio de Age por Survived y Pclass
promedio_sobrevivientes_por_clase = grouped_multi['Age'].mean()
promedio_sobrevivientes_por_clase
```

```python
# Desviación estándar de los sobrevivientes por clase
desviacion_estandar_sobrevivientes = df.groupby('Pclass')['Survived'].std()
desviacion_estandar_sobrevivientes
```
